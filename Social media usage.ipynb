```python
import pandas as pd
data = pd.read_csv('C:/Users/P3Fusion/Downloads/Time-Wasters on Social Media.csv')
data.head()
```

```python
data.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(data['Age'], bins=20, kde=True, color='blue', alpha=0.9)
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=data, hue='Gender', palette='Set2', legend=False)
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Platform', y='Addiction Level', data=data, hue='Platform', palette='coolwarm', legend=False)
plt.title('Addiction Level by Platform', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Addiction Level', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
numerical_features = data.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12, 8))
corr_matrix = data[numerical_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Total Time Spent', y='Engagement', hue='Platform', data=data, palette='viridis', alpha=0.7)
plt.title('Engagement vs. Total Time Spent', fontsize=16)
plt.xlabel('Total Time Spent (minutes)', fontsize=12)
plt.ylabel('Engagement', fontsize=12)
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(alpha=0.3)
plt.show()
```

```python
platform_counts = data['Video Category'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(platform_counts, labels=platform_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Video Category Distribution')
plt.show()
```
